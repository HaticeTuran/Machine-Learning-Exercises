3. Tackle the Titanic dataset. A great place to start is on Kaggle. Alternatively, you
can download the data from https://homl.info/titanic.tgz and unzip this tarball
like you did for the housing data in Chapter 2. This will give you two CSV files,
train.csv and test.csv, which you can load using pandas.read_csv(). The goal
is to train a classifier that can predict the Survived column based on the other
columns

```python
import pandas as pd
```

```python
test = pd.read_csv('test.csv')
test
```

```python
train = pd.read_csv('train.csv')
train
```

```python
train.info()
```

*   Most of the Cabin attributes are null. So we ignore this attribute.
*   Some of ages are also null. So we can fill the nulls with average age



```python
train = train.set_index("PassengerId")
test = test.set_index("PassengerId")
```

```python
train.info()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        # SimpleImputer: Replace missing values using a descriptive statistic (e.g. mean, median, or most frequent) along each column, or using a constant value.
        ("imputer", SimpleImputer(strategy="median")),
        # StandardScaler: Standardize features by removing the mean and scaling to unit variance.
        ("scaler", StandardScaler())
    ])
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

cat_pipeline = Pipeline([
        # OrdinalEncoder: Encode categorical features as an integer array.
        ("ordinal_encoder", OrdinalEncoder()), 
        # SimpleImputer: Replace missing values using a descriptive statistic (e.g. mean, median, or most frequent) along each column, or using a constant value.   
        ("imputer", SimpleImputer(strategy="most_frequent")),
        # Encode categorical features as a one-hot numeric array.
        #The input to this transformer should be an array-like of integers or strings, denoting the values taken on by categorical (discrete) features. 
        #The features are encoded using a one-hot (aka ‘one-of-K’ or ‘dummy’) encoding scheme. This creates a binary column for each category and returns a sparse matrix or dense array (depending on the sparse_output parameter)
        # When sparse = False then it  will return an array.(When true it returns sparse matrix)
        ("cat_encoder", OneHotEncoder(sparse=False)),
    ])
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = ["Age", "SibSp", "Parch", "Fare"]
cat_attribs = ["Pclass", "Sex", "Embarked"]

preprocess_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", cat_pipeline, cat_attribs),
    ])
```

```python
# fit_transform:  Fit the model and transform with the final estimator.
X_train = preprocess_pipeline.fit_transform(train)
X_train
```

```python
y_train = train["Survived"]
```

```python
from sklearn.ensemble import RandomForestClassifier

#RandomForestClassifier:  A random forest classifier.
# A random forest is a meta estimator that fits a number of decision tree classifiers on various sub-samples of the dataset 
# and uses averaging to improve the predictive accuracy and control over-fitting. 
forest_clf = RandomForestClassifier(n_estimators=100, random_state=42)
forest_clf.fit(X_train, y_train)
```

```python
X_test = preprocess_pipeline.transform(test)
y_pred = forest_clf.predict(X_test)
y_pred
```

```python
from sklearn.model_selection import cross_val_score

forest_scores = cross_val_score(forest_clf, X_train, y_train, cv=10)
forest_scores.mean()
```
